```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#To read the u.data file as a DataFrame called df_ratings:
ratings_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
df_ratings = pd.read_csv('ml-100K/u.data', sep='\t', names=ratings_cols)
print(df_ratings)
```

```python
#To get the General information about the u.data
print(df_ratings.info())
print('-----------------------------------------------------')
print(df_ratings.describe())
```

```python
#To read the u.item file as a DataFrame called df_movies:
movies_cols = ['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url']
df_movies = pd.read_csv('ml-100k/u.item', sep='|', encoding='latin-1', usecols=range(5), names=movies_cols)
print(df_movies)
```

```python
#To get the General information about the u.item
print(df_movies.info())
print('-----------------------------------------------------')
print(df_movies.describe())
```

```python
user_columns = ['user_id', 'age', 'gender', 'occupation', 'zip_code']
df_users = pd.read_csv('ml-100k/u.user', sep='|', names=user_columns)
print(df_users)
```

```python
#To get the General information about the u.item
print(df_users.info())
print('---------------------------------------------')
print(df_users.describe())
```

```python
#Use the head() method to show the first few rows of each DataFrame.
print(df_movies.head())
print('---------------------------------------------')
print(df_ratings.head())
print('---------------------------------------------')
print(df_users.head())
```

```python
#How many rows of data does each DataFrame have?
print("No. Of Rows: ",df_movies.shape[0])
print("No. Of Rows: ",df_ratings.shape[0])
print("No. Of Rows: ",df_users.shape[0])
```

```python
#Find the unique values and the number of unique values in the ‘rating’ column of the df_ratings.

unique_values = df_ratings['rating'].unique()
print("Unique Values in ratings Column are: ",unique_values)

count_values = df_ratings['rating'].nunique()
print("Number Of Unique values: ",count_values)
```

```python
#Find the number of unique values in the column 'user_id'. What does this number represent?

unique_user_ids = df_ratings['user_id'].nunique()
print("Number Of users Rating the movies: ",unique_user_ids)
#This Number indicates the Number of users giving ratings..
```

```python
#Show a histogram of the number of times each of the ratings were used.

plt.hist(df_ratings['rating'])
plt.title("Rating's plot")
plt.xlabel("Rating")
plt.show()
```

```python
#Which movie was rated the greatest number of times.

count_movie_rating = df_ratings['movie_id'].value_counts()
#print("Movie_id With greatest number of times",count_movie_rating.idxmax())
movie_title = df_movies[count_movie_rating.idxmax()==df_movies['movie_id']]['title']
print("Movie title for movie rated the greatest number of times: ",movie_title.values[0])
```

```python
#Find id of the movie that received the most '5' rating.
top_rated_movies = df_ratings[df_ratings['rating'] == 5]
most_5_rated_movie_id = top_rated_movies['movie_id'].value_counts().idxmax()
print("Movie ID for the Most 5 rated movie is: ",most_5_rated_movie_id)
```

```python
#Plot a histogram of the frequency of ratings by user.
user_rating_counts = df_ratings['user_id'].value_counts()
plt.hist(user_rating_counts, bins=30, edgecolor='black', alpha=0.7)
plt.xlabel('Number of Ratings by User')
plt.ylabel('Number of Users')
plt.title('Frequency of Ratings by User')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
#What is the date with the greatest number of movies released?
count_release_dates = df_movies['release_date'].value_counts()
print("Date with the greatest number of movies released is : ",count_release_dates.idxmax())
```

```python
#Merge the DataFrames for ratings and movies data for a comprehensive dataset. 
#Display a sample of rows from this combined DataFrame.

combined_dataset = pd.merge(df_ratings,df_movies,on='movie_id')
print(combined_dataset.sample(6))
```

```python
#How many times was the movie 'Star Wars (1977)' rated?

star_wars_movie_id=df_movies[df_movies['title']=='Star Wars (1977)']['movie_id'].values[0]
#print("Movie Id for Star Wars (1977): ",star_wars_movie_id)
star_wars_movie_rated_count = df_ratings[df_ratings['movie_id']==star_wars_movie_id]
print("The Movie 'Star Wars (1977)' is rated {} times".format(star_wars_movie_rated_count.shape[0]))
```

```python
#A Histogram using the number of ratings per movie
movie_rating = df_ratings['movie_id'].value_counts()

plt.hist(movie_rating.values)
plt.xlabel('Number Of Ratings')
plt.ylabel('Number Of Movies')
plt.title('Number of Ratings per Movie')
plt.show()
```

```python
#Show a bar chart of the top-rated movies.
movie_average_count=df_ratings.groupby(['movie_id'])['rating'].mean()

top_rated_movies = movie_average_count.index[:10]

plt.figure(figsize=(10,6))
plt.bar(top_rated_movies,movie_average_count[top_rated_movies])
plt.xlabel('Movie')
plt.ylabel('Average rating')
plt.title('Top Rated Movies')
plt.show()
```

```python
#Merge the user DataFrame with the previously merged DataFrame for users and movies data.
all_combined_dataset = pd.merge(df_users,combined_dataset,on='user_id')
print(all_combined_dataset)
```

```python
#Count the number of ratings and average rating for each movie

df = all_combined_dataset.groupby(['movie_id'])['rating'].agg(['count','mean'])
print("Number of ratings count and average rating for each movie:")
merged_df = pd.merge(df_movies,df,on='movie_id')
sorted_top_rated_movies = merged_df[['title', 'count','mean']].sort_values(by='count', ascending=False)
print(sorted_top_rated_movies)
```

```python
#Find the top-rated movies with a minimum rating count of 100.
min_100_rating_count = all_combined_dataset.groupby(['movie_id'])['rating'].agg(['count'])
top_rated_movies = min_100_rating_count[min_100_rating_count>=100].dropna()
merged_df = df_movies.merge(top_rated_movies, left_on='movie_id', right_index=True)
sorted_top_rated_movies = merged_df[['movie_id', 'title', 'count']].sort_values(by='count', ascending=False)
print(sorted_top_rated_movies.head(10))
```

```python
#Find the average rating for each movie by gender. Draw a histogram for each gender
avg_rate_gender = all_combined_dataset.groupby(['movie_id','gender'])['rating'].mean().unstack()
#print(avg_rate_gender)

#plt.figure(figsize=(10,6))
genders = ['M', 'F']
for gender in genders:
    plt.hist(avg_rate_gender[gender])
    plt.xlabel("Average Rating")
    plt.ylabel("Number of Movies")
    plt.title("Movie Ratings for {} Gender".format(gender))
    plt.show()
```

```python
#Give a list of the number of people in each occupation in descending order.
occupation_counts = df_users['occupation'].value_counts().reset_index()
occupation_counts.columns = ['Occupation', 'Number of People']
occupation_counts = occupation_counts.sort_values(by='Number of People', ascending=False)
print(occupation_counts)
```

```python
#Plot a bar chart of occupations.
occupation_counts=occupation_counts.set_index('Occupation')
plt.figure(figsize=(12, 6))
occupation_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Occupation')
plt.ylabel('Number of People')
plt.title('Bar Chart For Occupations')
plt.show()
```
